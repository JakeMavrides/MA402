```python
import numpy as np
import matplotlib.pyplot as plt
import math
'''
Edge length now given by S(n) = S(n-1) / sqrt(2+sqrt(4-[S(n-1)]^2))
  S(1) = 1 ; n>=2
'''
# Question 1
def C(n):
  return E(n)*S(n)

def E(n):
  return 3 * (2**n)

def S(n):
  if (n == 1):
    return 1
  else:
    prev = S(n-1) #Helps runtime 
    innerRoot = math.sqrt(4 - (prev)**2)
    return prev / math.sqrt(2 + innerRoot)

def absError(Cn):
  return abs(Cn - (2 * math.pi))

xlist = np.linspace(1, 30, num=30)
Cvals = [] #C(n) values
Evals = [] #Abs. error values
for i in range(1, 31):
  curr = C(i)
  Evals.append(absError(curr))
  Cvals.append(curr)

#Plot #1 - abs error on semilogy
plt.subplot(211)
plt.semilogy(xlist, Evals)
plt.xlabel("n-values")
plt.ylabel("log of absolute error")
plt.title("Absolute error on semilogy -- Question 1")

plt.subplot(212)
plt.plot(xlist, Cvals)
plt.xlabel("n-values")
plt.ylabel("C(n)")
plt.title("2Ï€ approximation values - visual")

plt.tight_layout() # Spaces the graphs nicely
plt.show()
```
